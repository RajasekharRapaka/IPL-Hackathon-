```python
# Winner by toss decision
```

```python
# Calculate the number of matches won by teams based on toss decision
matches_won_by_fielding = matches[matches['toss_decision'] == 'field']['winner'].value_counts()
matches_won_by_batting = matches[matches['toss_decision'] == 'bat']['winner'].value_counts()

# Calculate the total matches won by each team
total_matches_won = matches['winner'].value_counts()

# Calculate the winning percentage for each decision
percentage_won_by_fielding = matches_won_by_fielding / total_matches_won
percentage_won_by_batting = matches_won_by_batting / total_matches_won

# Display the results
print("Percentage of matches won by choosing to field first:")
print(percentage_won_by_fielding)

print("\nPercentage of matches won by choosing to bat first:")
print(percentage_won_by_batting)
```

```python
# Strike rate of Kholi in 2016
```

```python
# Filter data for Kohli in 2016
kohli_2016 = ipl[(ipl['batsman'] == 'V Kohli') & (ipl['season'] == 2016)]

# Calculate the total runs and total balls faced
total_runs = kohli_2016['total_runs'].sum()
total_balls_faced = kohli_2016['ball'].sum()

# Calculate the strike rate
strike_rate = (total_runs / total_balls_faced) * 100

print(f"The strike rate of Kohli in 2016 is: {strike_rate:.2f}")
```
